````python
%cd /content/DeepSpeech
!python3 DeepSpeech.py --train_cudnn True --early_stop True --es_epochs 6 --n_hidden 2048 --epochs 30 \
  --export_dir /content/models/ --checkpoint_dir /content/deepspeech-0.9.3-checkpoint/ \
  --train_files /content/drive/MyDrive/csvs/train.csv --dev_files /content/drive/MyDrive/csvs/dev.csv --test_files /content/drive/MyDrive/csvs/test.csv \
  --learning_rate 0.00001 --train_batch_size 1 --test_batch_size 1 --dev_batch_size 1 --export_file_name 'ft_model' --verbosity=1 \
  --augment reverb[p=0.5,delay=50.0~30.0,decay=10.0:2.0~1.0] \
  --augment volume[p=0.5,dbfs=-10:-40] \
  --augment pitch[p=0.5,pitch=1~0.2] \
  --augment tempo[p=0.5,factor=1~0.5] \
  --reduce_lr_on_plateau true \
  --plateau_epochs 5 \
  --plateau_reduction 0.1 \
  --export_tflite
````
