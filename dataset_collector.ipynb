```python
import sounddevice as sd
import numpy as np
import scipy
import shutil
import os
from scipy.io.wavfile import write

commands = [
    'Catch phrase',
    'Noise',
    'command3',
]

def is_this_integer(value):
    try:
        number = int(value)
        return True
    except ValueError:
        return False
    
def is_this_in_the_range(value):
    val = int(value)
    if not(1 <= val <= len(commands)):
        return False
    return True
    

    
print("In this program, you will record your audio for every command 100 times! Sorry, but we need this data for our project.")
print("For which command are you trying to record your audio?")
print("Choose the command from the list below and then just write down the number: ")
for i, com in enumerate(commands):
    print(f"{i+1}. {com}")
    
    
val = input("enter a number: ")
while not(is_this_integer(val)) or not(is_this_in_the_range(val)):
    if not(is_this_integer(val)):
        print("Please enter a number.")
        val = input()
    elif not(is_this_in_the_range(val)):
        print(f"Please enter a number from 1 to {len(commands)}")
        val = input()

    
##### Removing the directory if it already exists (might happen when running the script for the second time) ######

if os.path.exists(f"{commands[int(val)-1]}"):
    path = os.path.join(os.getcwd(), f"{commands[int(val)-1]}")
    shutil.rmtree(path)

### recording audio files ###
i = 0
number_of_files_needed = 100
while i < number_of_files_needed:
  
    # Sampling frequency
    freq = 44100

    # Recording duration
    duration = 3.5

    # Start recorder with the given values 
    # of duration and sample frequency
    recording = sd.rec(int(duration * freq), 
                       samplerate=freq, channels=2)

    # Record audio for the given number of seconds
    sd.wait()
##################################
    confirmation = input(f"File #{i+1} Do you want to save it? 1 for yes and 0 for no: ")
    while not(is_this_integer(confirmation)) or not(int(confirmation) == 0 or int(confirmation) == 1):
        if not(is_this_integer(confirmation)):
            print("Please enter a number.")
            confirmation = input()
        elif not(int(confirmation) == 0 or int(confirmation) == 1):
            print("0 or 1, nothing else!")
            confirmation = input()
    
    # if no then i = i -1
    # This will convert the NumPy array to an audio
    # file with the given sampling frequency
    confirmation_ = int(confirmation)
    if confirmation_ == 1:
        write(f"{commands[int(val)-1]}-{i}.wav", freq, recording)
        if not os.path.exists(f"{commands[int(val)-1]}"):
            os.mkdir(f"{commands[int(val)-1]}")
        try:
            source = os.getcwd() + f"\\{commands[int(val)-1]}-{i}.wav"
            dest = os.getcwd() + f"\\{commands[int(val)-1]}"
            shutil.move(source, dest)
        except:
            print("Please move all the audio files into a zip file")
    else:
        i -= 1
##############################
#     write(f"{commands[int(val)-1]}-{i+300}.wav", freq, recording)
    i += 1

try:   
    shutil.make_archive(f"{commands[int(val)-1]}s", "zip", f"{commands[int(val)-1]}")
except:
    pass

print("Done!")
```
